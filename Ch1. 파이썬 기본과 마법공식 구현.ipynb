```python
#[코드 1.5] 마법공식 데이터 파일 연결하기 (Ch1. 파이썬 기본과 마법공식 구현.ipynb)

file_path = '/Users/chisunjoung/quantpython-master/주식 책 데이터/마법공식 데이터.xlsx'
wb = xlrd.open_workbook(file_path)
```

```python
#[코드 1.6] PER시트와 연결하기(Ch1. 파이썬 기본과 마법공식 구현.ipynb)

per_sh = wb.sheet_by_name('PER')
```

```python
#[코드 1.7] PER시트에서 행 별로 데이터 읽어와 저장하기 (Ch1. 파이썬 기본과 마법공식 구현.ipynb)

per_dict = { }
for i in range(1, per_sh.nrows):  
    data = per_sh.row_values(i)
    name = data[0]
    per = data[1]
    if per > 0:
        per_dict[name] = per
```

```python
#[코드 1.8] 저장한 PER딕셔너리 정렬하기 (Ch1. 파이썬 기본과 마법공식 구현.ipynb)

import operator

sorted_per = sorted(per_dict.items(), key=operator.itemgetter(1))
```

```python
#[코드 1.9] PER랭크 메기기 (Ch1. 파이썬 기본과 마법공식 구현.ipynb)

per_rank = {}

for num, firm in enumerate(sorted_per):
    per_rank[firm[0]] = num + 1 
```

```python
#[코드 1.11] ROA시트에서 행 별로 데이터 읽어와 저장하기 (Ch1. 파이썬 기본과 마법공식 구현.ipynb)

roa_dict = { }

for i in range(1, roa_sh.nrows):  
    data = roa_sh.row_values(i)
    name = data[0]
    roa = data[1]
    if roa != '':
        roa_dict[name] = roa
```

```python
#[코드 1.12] 저장한 ROA딕셔너리 정렬하기 (Ch1. 파이썬 기본과 마법공식 구현.ipynb)

sorted_roa = sorted(roa_dict.items(), key=operator.itemgetter(1), reverse=True)
```

```python
#[코드 1.17] 마법공식 코드 정리 (Ch1. 파이썬 기본과 마법공식 구현.ipynb)

import xlrd 
import operator

file_path = 'C:\\Users\\JK\\Desktop\\파이썬으로 하는 퀀트투자\\마법공식 데이터.xlsx'
wb = xlrd.open_workbook(file_path)

per_sh = wb.sheet_by_name('PER')

per_dict = { }
for i in range(1, per_sh.nrows):  
    data = per_sh.row_values(i)
    name = data[0]
    per = data[1]
    if per > 0:
        per_dict[name] = per
        
sorted_per = sorted(per_dict.items(), key=operator.itemgetter(1))

per_rank = {}
for num, firm in enumerate(sorted_per):
    per_rank[firm[0]] = num + 1 
    
roa_sh = wb.sheet_by_name('ROA')

roa_dict = { }
for i in range(1, roa_sh.nrows):  
    data = roa_sh.row_values(i)
    name = data[0]
    roa = data[1]
    if roa != '':
        roa_dict[name] = roa
        
sorted_roa = sorted(roa_dict.items(), key=operator.itemgetter(1), reverse=True)

roa_rank = {}
for num, firm in enumerate(sorted_roa):
    roa_rank[firm[0]] = num + 1
    
total_rank = {}
for name in roa_rank.keys():
    if name in per_rank.keys():
        total_rank[name] = per_rank[name] + roa_rank[name]
        
sorted_total = sorted(total_rank.items(), key=operator.itemgetter(1))

magic_rank = {}
for num, firm in enumerate(sorted_total):
    magic_rank[firm[0]] = num + 1
```

```python
# [코드 1.24] 마법공식 함수화(Ch1. 파이썬 기본과 마법공식 구현.ipynb)

def magic(file_path):
    wb = xlrd.open_workbook(file_path)

    per_sh = wb.sheet_by_name('PER')

    per_dict = { }
    for i in range(1, per_sh.nrows):  
        data = per_sh.row_values(i)
        name = data[0]
        per = data[1]
        if per > 0:
            per_dict[name] = per

    sorted_per = sorted(per_dict.items(), key=operator.itemgetter(1))

    per_rank = {}
    for num, firm in enumerate(sorted_per):
        per_rank[firm[0]] = num + 1 

    roa_sh = wb.sheet_by_name('ROA')

    roa_dict = { }
    for i in range(1, roa_sh.nrows):  
        data = roa_sh.row_values(i)
        name = data[0]
        roa = data[1]
        if roa != '':
            roa_dict[name] = roa

    sorted_roa = sorted(roa_dict.items(), key=operator.itemgetter(1), reverse=True)

    roa_rank = {}
    for num, firm in enumerate(sorted_roa):
        roa_rank[firm[0]] = num + 1

    total_rank = {}
    for name in roa_rank.keys():
        if name in per_rank.keys():
            total_rank[name] = per_rank[name] + roa_rank[name]

    sorted_total = sorted(total_rank.items(), key=operator.itemgetter(1))

    magic_rank = {}
    for num, firm in enumerate(sorted_total):
        magic_rank[firm[0]] = num + 1
        
    return magic_rank
```
